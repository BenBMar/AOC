```python
import re

with open('day02.txt' , 'r') as file :
     input= file.read()
comma_numbers = input.split(',')
print (comma_numbers)

 
undashed_numbers_full = []
for i in range (len(comma_numbers)):
    temp_list = [] 
    num1_strc , num2_strc = comma_numbers[i].split('-')
    num1 = int (num1_strc)      
    num2 = int (num2_strc)
    temp_list = list(range(num1, num2 + 1))
    undashed_numbers_full.append(temp_list)
    
undashed_numbers = [item for sublist in undashed_numbers_full for item in sublist]


def is_twice(n):
    s = str(n)
    length = len(s)
    if length % 2 != 0:
        return False    
    midpoint = length // 2
    first_half = s[:midpoint]           
    second_half = s[midpoint:]           
    return first_half == second_half

twice_numbers = [num for num in undashed_numbers if is_twice(num)]
print (twice_numbers)
print(sum(twice_numbers))

```
